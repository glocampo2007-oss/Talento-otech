### Actividad 1: Juego de adivinar el número

Aquí tienes el código para cada uno de los pasos que solicitaste para crear el juego de adivinar el número.

#### Paso 1 y 2: Importar el módulo `random` y crear la función `generar_numero()`

```python
import random

def generar_numero():
    """Genera un número entero aleatorio entre 1 y 100."""
    return random.randint(1, 100)

print("Función generar_numero() creada.")
```

#### Paso 3, 4, 5, 6 y 7: Implementar el bucle principal del juego con lógica, manejo de errores, pistas y contador.

```python
def jugar_adivina_numero():
    numero_secreto = generar_numero()
    intentos = 0
    adivinado = False

    print("¡Bienvenido al juego de adivinar el número!")
    print("He pensado en un número entre 1 y 100. ¡Intenta adivinarlo!")

    while not adivinado:
        try:
            intento_jugador = int(input("Introduce tu suposición: "))
            intentos += 1

            if intento_jugador < 1 or intento_jugador > 100:
                print("Por favor, introduce un número entre 1 y 100.")
            elif intento_jugador < numero_secreto:
                print("El número secreto es mayor. ¡Sigue intentando!")
            elif intento_jugador > numero_secreto:
                print("El número secreto es menor. ¡Sigue intentando!")
            else:
                print(f"¡Felicidades! Has adivinado el número {numero_secreto} en {intentos} intentos.")
                adivinado = True
        except ValueError:
            print("Entrada no válida. Por favor, introduce un número entero.")

# Puedes llamar a la función para jugar:
jugar_adivina_numero()
```

### Actividad 2: Calculadora de operaciones básicas

Aquí tienes el código para crear una calculadora simple con las operaciones básicas.

#### Paso 1: Definir funciones para cada operación (suma, resta, multiplicación, división)

```python
def sumar(a, b):
    return a + b

def restar(a, b):
    return a - b

def multiplicar(a, b):
    return a * b

def dividir(a, b):
    if b == 0:
        raise ValueError("No se puede dividir por cero.")
    return a / b

print("Funciones de operación definidas.")
```

#### Paso 2 al 6: Implementar el bucle principal de la calculadora con menú, selección de operación, manejo de errores, y opción para salir.

```python
def calculadora():
    while True:
        print("\n--- Calculadora Simple ---")
        print("1. Sumar")
        print("2. Restar")
        print("3. Multiplicar")
        print("4. Dividir")
        print("5. Salir")

        opcion = input("Selecciona una opción (1-5): ")

        if opcion == '5':
            print("¡Gracias por usar la calculadora!\nAdiós.")
            break

        if opcion in ('1', '2', '3', '4'):
            try:
                num1 = float(input("Introduce el primer número: "))
                num2 = float(input("Introduce el segundo número: "))
            except ValueError:
                print("Entrada no válida. Por favor, introduce números.")
                continue

            try:
                if opcion == '1':
                    print(f"Resultado: {num1} + {num2} = {sumar(num1, num2)}")
                elif opcion == '2':
                    print(f"Resultado: {num1} - {num2} = {restar(num1, num2)}")
                elif opcion == '3':
                    print(f"Resultado: {num1} * {num2} = {multiplicar(num1, num2)}")
                elif opcion == '4':
                    print(f"Resultado: {num1} / {num2} = {dividir(num1, num2)}")
            except ValueError as e:
                print(f"Error: {e}")
        else:
            print("Opción no válida. Por favor, selecciona un número entre 1 y 5.")

# Puedes llamar a la función para usar la calculadora:
calculadora()
```

### Actividad 3: Lista de tareas

Aquí tienes el código para crear una aplicación de lista de tareas.

#### Paso 1 y 2: Crear una lista vacía para almacenar las tareas y definir funciones para agregar, eliminar, y mostrar tareas.

```python
tareas = [] # Lista global para almacenar las tareas

def agregar_tarea(tarea):
    """Agrega una tarea a la lista."""
    tareas.append(tarea)
    print(f"Tarea '{tarea}' agregada.")

def eliminar_tarea(indice):
    """Elimina una tarea de la lista por su índice."""
    try:
        tarea_eliminada = tareas.pop(indice - 1) # Restar 1 porque los índices son base 0
        print(f"Tarea '{tarea_eliminada}' eliminada.")
    except IndexError:
        print("Error: El número de tarea no es válido.")

def mostrar_tareas():
    """Muestra todas las tareas en la lista."""
    if not tareas:
        print("No hay tareas en la lista.")
    else:
        print("\n--- Tus Tareas ---")
        for i, tarea in enumerate(tareas):
            print(f"{i + 1}. {tarea}") # Sumar 1 para mostrar números de tarea base 1
        print("-------------------")

print("Funciones de gestión de tareas definidas.")
```

#### Paso 3, 4, 5, 6 y 7: Implementar un menú principal usando un bucle while, manejar opciones con if-elif-else, mostrar tareas con bucle for, manejar errores y opción para salir.

```python
def lista_de_tareas():
    while True:
        print("\n--- Menú de Lista de Tareas ---")
        print("1. Agregar tarea")
        print("2. Eliminar tarea")
        print("3. Mostrar tareas")
        print("4. Salir")

        opcion = input("Selecciona una opción (1-4): ")

        if opcion == '1':
            tarea = input("Introduce la nueva tarea: ")
            agregar_tarea(tarea)
        elif opcion == '2':
            mostrar_tareas()
            try:
                indice = int(input("Introduce el número de la tarea a eliminar: "))
                eliminar_tarea(indice)
            except ValueError:
                print("Entrada no válida. Por favor, introduce un número entero.")
        elif opcion == '3':
            mostrar_tareas()
        elif opcion == '4':
            print("¡Gracias por usar la Lista de Tareas! Adiós.")
            break
        else:
            print("Opción no válida. Por favor, selecciona un número entre 1 y 4.")

# Puedes llamar a la función para usar la lista de tareas:
lista_de_tareas()
```
